```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(units=1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
history = model.fit(
    celsius,
    fahrenheit,
    epochs=10000,
    verbose=0
)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

```python
c_value = 100.0
predicted_f = model.predict(np.array([[c_value]]))[0][0]
actual_f = (c_value * 9/5) + 32

print(f"Celsius: {c_value}°C")
print(f"Predicted Fahrenheit (ANN): {predicted_f:.2f}°F")
print(f"Actual Fahrenheit (Formula): {actual_f:.2f}°F")
```
